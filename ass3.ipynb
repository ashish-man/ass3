```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
sns.barplot(df.sex,df.age)
```

```python
sns.barplot(df.age,df.cp.sort_values() )
```

```python
df.groupby('cp').cp.count()
```

```python
a,b,c,d=df.groupby('cp').cp.count()
```

```python
labels = 'pain_level-1', 'pain_level-2', 'pain_level-3', 'pain_level-4'
sizes = [a, b,c, d]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) 

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
